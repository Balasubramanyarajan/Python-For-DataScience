```python
import matplotlib.pyplot as plt
import numpy as np
```

Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Scatter Plot

```python
x = [5, 7, 8, 10, 12]
y = [10, 14, 15, 18, 20]

plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

Bar Chart

```python
x = ['Apple', 'Banana', 'Cherry']
y = [30, 45, 10]

plt.bar(x, y, color='orange')
plt.title("Fruit Sales")
plt.xlabel("Fruit")
plt.ylabel("Quantity Sold")
plt.show()
```

Horizontal Bar Chart

```python
x = ['Python', 'Java', 'C++']
y = [90, 75, 60]

plt.barh(x, y, color='skyblue')
plt.title("Programming Language Popularity")
plt.xlabel("Score")
plt.ylabel("Language")
plt.show()
```

Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, color='green', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

Pie Chart

```python
sizes = [40, 35, 25]
labels = ['Python', 'Java', 'C++']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Language Popularity Share")
plt.show()
```

Box Plot

```python
data = [5, 8, 10, 15, 18, 20, 22, 25, 30]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

Subplot

```python
x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 3, 5, 7]

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2)
plt.title("Plot 2")

plt.show()
```

Legend

```python
x = [1, 2, 3, 4]
plt.plot(x, [1, 4, 9, 16], label='Square')
plt.plot(x, [1, 8, 27, 64], label='Cube')

plt.legend()
plt.title("Legend Example")
plt.show()
```

Grid

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.grid(True)
plt.title("Line Plot with Grid")
plt.show()
```

Figure

```python
plt.figure(figsize=(6,4))
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Custom Figure Size")
plt.show()
```

Save Figure

```python
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Save Plot Example")
plt.savefig("plot_output.png")
plt.show()
```

Error Bar

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
error = [0.2, 0.4, 0.3, 0.5]

plt.errorbar(x, y, yerr=error, fmt='o', color='black', ecolor='red', capsize=5)
plt.title("Error Bar Example")
plt.show()
```

Fill Between

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.title("Fill Between Example")
plt.show()
```

Annotate

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.annotate("Peak Point", xy=(5, 10), xytext=(3, 9),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```
